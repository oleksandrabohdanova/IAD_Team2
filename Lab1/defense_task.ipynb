##### **Навести приклад створення структурованого масиву в NumPy.**

Структуровані масиви – це масиви типу ndarray, тип даних яких є композицією простіших типів даних, організованих у вигляді послідовності іменованих полів. Наприклад,

```python
import numpy as np
titanic_dtype = np.dtype([
    ('name', 'U20'),     
    ('sex', 'U6'),       
    ('age', 'i4'),       
    ('pclass', 'i1'),    
    ('survived', 'i1')   
])
data = np.array([
    ('Braund, Mr. Owen', 'male',   22, 3, 0),
    ('Cumings, Mrs. J.', 'female', 38, 1, 1),
    ('Heikkinen, Miss. L.', 'female', 26, 3, 1),
    ('Futrelle, Mrs. J.', 'female', 35, 1, 1),
    ('Allen, Mr. W.',    'male',   35, 1, 0)
], dtype=titanic_dtype)
```

Тут data — це одновимірний масив довжини 2, чий тип (dtype) є структурою з п’ятьма полями:

> 1. name — рядок Unicode довжиною до 20 символів.
> 2. sex — рядок Unicode довжиною до 6 символів (male/female).
> 3. age — 32-бітне ціле число (вік пасажира).
> 4. pclass — 8-бітне ціле число (номер класу квитка).
> 5. survived — 8-бітне ціле число (0 = не вижив, 1 = вижив).

Якщо зробимо виклик data[:3], отримаємо підмасив з перших трьох структур:

```python
print("Перші записи:")
print(data[:3])
```

Індексуючи по імені поле ('name'), ми можемо отримати всі його значення як окремий масив:

```python
print("\nІмена всіх пасажирів:", data['name'])
```

Над полями також можемо виконувати звичайні векторизовані операції, як наприклад, отримання середнього значення:

```python
print("\nСередній вік:", data['age'].mean())
```

Окрім цього, можемо створювати булевий масив, який відповідає кільком вимогам. Наприклад, повернути всі значення, де пасажири жінки та з 1 класу:

```python
print("\nЖінки 1-го класу:")
print(data[(data['sex'] == 'female') & (data['pclass'] == 1)])
```
